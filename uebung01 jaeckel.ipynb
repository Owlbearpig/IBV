c) Schreiben Sie eine Funktion, die die n größten Werte in einem Numpy Array berechnet und messen Sie die Laufzeit hierfür in IPython mit `%%time`.

```python
%%time
import numpy as np

def n_highest_numbers(n, arr):
    arr = arr.flatten()        
    arr = np.sort(arr)
    if n <= arr.shape[0]:
        return arr[-n:]

arr = np.random.uniform(size=(100, 100, 100))
n = 500

print(n_highest_numbers(n, arr))
```

b) Der Hamming-Abstand von Binärcodes ist definiert als $ h(x, y) := | \{j ∈ \{1, ... , l\} | x_j  \neq y_j \} |$ wobei $ j \in {0, 1} $. Schreiben Sie ein Programm, das zunächst zwei zufällige Binärcodes der Länge l generiert und anschließend ihren Hamming-Abstand berechnet.
*Hinweis:* Benutzen Sie die Funktion `zip`. Eine genaue Beschreibung der Funktion erhält man durch `print zip.__doc__`

```python
l = 10
x = np.random.randint(2, size=l)
y = np.random.randint(2, size=l)
print(x)
print(y)
d = 0
for tuple in zip(x,y):
    if (tuple[0] != tuple[1]):
        d += 1
print(d)
```

#### Aufgabe 1.3: Higher Order Functions (2 Punkte)

Schreiben Sie ein Programm, das eine Liste von Wörtern auf eine Liste von Zahlen abbildet, wobei die Zahlen die Länge der Wörter repräsentieren. Schreiben Sie das Programm in drei Varianten: 1) mit einer for-Schleife, 2) mit der Funktion `map()` und 3) mit List Comprehensions.

```python
word_list = ["orange", "grape", "strawberry", "avocado", "huckleberry"]

# 1)
def word_map(word_list):
    ret = []
    for word in word_list:
            ret.append(len(word))
    return ret

print("1)", word_map(word_list))

# 2)
print("2)", list(map(len, word_list)))

# 3)
print("3)", [len(word) for word in word_list])

```

#### Aufgabe 2.2: Bildmanipulation mit PIL (1+1+1 Punkte)
Schreiben Sie ein Programm, das die folgenden Bearbeitungsschritte auf dem in Aufgabe 2.1 heruntergeladenen Bild durchführt und implementieren Sie entsprechende Funktionen. Dazu sollte die Python Image Library (PIL) verwendet werden.

a) Schneiden Sie einen 45 x 52 Pixel großen Ausschnitt aus dem Bild heraus, wobei die linke obere Ecke des Ausschnitts einen Abstand von (335,18) von der des Bildes hat. Geben Sie diesen Ausschnitt im IPython Notebook aus.

```python
from PIL import ImageOps
img_cut = img.crop((335, 18, 335+45, 18+52))
plt.imshow(img_cut)
```

b) Das ausgeschnittene Bild soll nun vertikal gespiegelt und in die Mitte eines neuen, 100 x 100 Pixel großen Bildes plaziert und anschließend angezeigt werden.

```python
img_cut_mirror = ImageOps.mirror(img_cut)
img_new = PIL.Image.new("RGB", (100, 100))
img_new.paste(img_cut_mirror, ((100 - img_cut_mirror.size[0])//2, (100 - img_cut_mirror.size[1])//2))

plt.imshow(img_new)
```

c) Zeichnen Sie an beliebiger Stelle im Bild das Wort “PIL” ein, zeigen Sie das Bild an und speichern es anschließend auf der Festplatte als PNG.

```python
from PIL import ImageDraw, ImageFont
draw = ImageDraw.Draw(img_new)

draw.text((img_new.size[0]//2, img_new.size[1]//2), "PIL")

plt.imshow(img_new)

img_new.save("final.PNG")
```
